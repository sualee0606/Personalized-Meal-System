```python
import pandas as pd

df = pd.read_csv('FOOD-DATA.csv')

non_vegetarian_keywords = [
    "chicken", "beef", "pork", "lamb", "fish", "shrimp", "crab", "milk", 
    "cheese", "butter", "yogurt", "egg", "meat", "bacon", "sausage", "ham"
]

vegetarian_df = df[~df['food'].str.lower().apply(lambda x: any(keyword in x for keyword in non_vegetarian_keywords))]

vegan_keywords = [
    "tofu", "tempeh", "lentil", "chickpea", "beans", "quinoa", "oats", 
    "rice", "cucumber", "tomato", "spinach", "broccoli", "carrot", "avocado", 
    "garlic", "onion", "olive oil", "coconut oil", "avocado oil", "seeds", 
    "nuts", "almond", "cashew", "walnut", "soy", "fruit", "herb"
]

vegan_df = vegetarian_df[vegetarian_df['food'].str.lower().apply(lambda x: any(keyword in x for keyword in vegan_keywords))]

print(vegan_df[['food', 'Caloric Value', 'Protein']])

vegan_df.to_csv('FOOD-VEGAN-DEMO.csv', index=False)
```

```python
import pandas as pd

vegan_foods = pd.read_csv('FOOD-VEGAN-DEMO.csv')

exclude_foods = [
    "Chicken and rice casserole homemade",
    "Baked potato with cheese sauce bacon",
    "Corned beef hash with potatoes",
    "Sardines in tomato sauce canned",
    "Pasta with sliced franks in tomato sauce",
    "Cheese pizza",
    "Butter cookies",
    "Milkshake",
    "French toast",
    "Chocolate bars with milk",
    "Onion rings burger king",
    "Tomato soup canned",
    "Peanuts chocolate coated",
    "Vanilla fudge with nuts",
    "Fruitcake",
    "Danish pastry with nuts",
    "Fruit juice bar",
    "Rice pudding",
    "Caramel med iced coffee Dunkin Donuts"
]

vegan_foods['food_clean'] = vegan_foods['food'].str.strip().str.lower()
exclude_foods_clean = [food.lower().strip() for food in exclude_foods]

vegan_foods_filtered = vegan_foods[~vegan_foods['food_clean'].isin(exclude_foods_clean)]

vegan_foods_filtered = vegan_foods_filtered.drop(columns=['food_clean'])

vegan_foods_filtered.to_csv('filtered_vegan_foods.csv', index=False)

print("Filtered vegan foods:")
print(vegan_foods_filtered)
```

```python
import pandas as pd

df = pd.read_csv("FOOD-DATA.csv")

target_foods = [
    "chicken breast",
    "egg",
    "beef tenderloin",
    "pork tenderloin",
    "salmon",
    "tuna",
    "cod",
    "lentils",
    "chickpeas",
    "black beans",
    "tofu",
    "tempeh",
    "greek yogurt",
    "cottage cheese",
    "protein powder"
]

df["food_lower"] = df["food"].str.lower()

matched_results = pd.DataFrame()

for food in target_foods:
    pattern = food.lower()
    matches = df[df["food_lower"].str.contains(pattern, na=False)]
    if not matches.empty:
        print(f"Found matches for '{food}':")
        print(matches[["food", "Caloric Value", "Protein"]])
        print("-" * 50)
        matched_results = pd.concat([matched_results, matches[["food", "Caloric Value", "Protein"]]], ignore_index=True)
    else:
        print(f"No matches found for '{food}'.")

matched_results = matched_results.drop_duplicates()

if not matched_results.empty:
    matched_results.to_csv("FFOOD-HIGH-PROTEIN.csv", index=False)
    print("Matched results saved to 'FOOD-HIGH-PROTEIN.csv'.")
else:
    print("No matched results found. No CSV file created.")
```
